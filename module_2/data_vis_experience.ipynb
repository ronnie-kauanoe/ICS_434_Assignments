#### 2. Generating a scatter plot
* Generate random data to produce a scatter plot
  * You can use any function to generate the data
  * Instead of using circles as symbols, you should use squares.
    

```python
# Set random seed to 42
np.random.seed(42)
# Total number of points to generate
count = 1000
# Randomly generate x-axis values
X = [(np.random.random() * 100) for _ in range(count)]
# Randomly generate y-axis values
Y = [(np.random.random() * 100) for _ in range(count)]
# Randomly generate colors
colors = ["#%06x" % np.random.randint(0, 0xFFFFFF) for _ in range(count)]
# Randomly generate point colors
sizes = [(np.random.random() * 500) for _ in range(count)]
# Set chart dimensions
plt.figure(figsize=(12, 8), dpi=80)
# Feed x-values, y-values, colors list, and sizes list into scatter plot
# Change marker to "s" for square markers
plt.scatter(X, Y, c = colors, s = sizes, marker = "s")
# Add title
plt.title("Random Scatter Plot")
# Add labels to axes
plt.xlabel("Random X")
plt.ylabel("Random Y")
# Display the graph
plt.show()
```
